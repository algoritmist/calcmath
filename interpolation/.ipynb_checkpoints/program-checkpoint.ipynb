```python
from interpolation.lagrange.solver import *
from generators.generator import generate_data
```

```python
test_function = lambda x: -x ** 3 + 2.5 * x ** 2 + 1.4 * x - 0.6
left_range, right_range = -1, 3

import matplotlib.pyplot as plt
from interpolation.generators.generator import generate_range

nodes = 20
f_nodes = 100
data = generate_data(test_function, left_range, right_range, nodes)
print(data)
t = generate_range(left_range, right_range, f_nodes)
plt.figure()
plt.plot(t, test_function(t), 'r')
plt.show()
```

```python
x_data = list(map(lambda t: t[0], data))
y_data = list(map(lambda t: t[1], data))

lagrange_polynomial = create_lagrange_polynomial(x_data, y_data)

t1 = generate_range(left_range, right_range, nodes)
t2 = generate_range(left_range, right_range, f_nodes)

plt.figure()
plt.plot(t1, lagrange_polynomial(t1), 'b')
plt.show()
plt.figure()
plt.plot(t2, test_function(t2), 'b')
plt.show()
```

```python
import numpy as np


def generate_noise_data(f, a, b, nodes):
    return [(a - 1, -f(a - 1) * 3)] + [(x, f(x)) for i, x in enumerate(generate_range(a, b, nodes))] + [
        (b + 1, -f(b + 1) * 3)]


def generate_noise_range(a, b, nodes):
    return np.linspace(a - 1, b + 1, nodes)


noise_data = generate_noise_data(test_function, left_range, right_range, nodes)
x_noise = list(map(lambda g: g[0], noise_data))
y_noise = list(map(lambda g: g[1], noise_data))

lagrange_polynomial_noise = create_lagrange_polynomial(x_noise, y_noise)

q = generate_noise_range(left_range, right_range, nodes)
_q = generate_noise_range(left_range, right_range, 10 * nodes)

plt.figure()
plt.plot(q, lagrange_polynomial_noise(q), 'bo', _q, lagrange_polynomial_noise(_q))
plt.show()
plt.figure()
plt.plot(q, test_function(q), 'r')
plt.show()
```

```python
import math


def generate_chebyshev_range(a, b, nodes):
    x = []
    for i in range(nodes + 1):
        ksi = -math.cos((2 * i + 1) / (2 * nodes + 2) * math.pi)
        x.append((a + b) / 2 + (b - a) / 2 * ksi)
    return x


q = generate_chebyshev_range(left_range, right_range, nodes)
_q = generate_chebyshev_range(left_range - 1, right_range + 1, 10 * nodes)

chebyshev_data = [(left_range - 1, -test_function(left_range - 1) * 3)] \
                 + [(x, test_function(x)) for x in q] + \
                 [(right_range + 1, -test_function(right_range + 1) * 3)]
x_chebyshev = list(map(lambda g: g[0], chebyshev_data))
y_chebyshev = list(map(lambda g: g[1], chebyshev_data))
print(x_chebyshev)
print(y_chebyshev)

lagrange_polynomial_chebyshev = create_lagrange_polynomial(x_chebyshev, y_chebyshev)

y_q = [lagrange_polynomial_chebyshev(x) for x in q]
y__q = [lagrange_polynomial_chebyshev(x) for x in _q]

plt.figure()
plt.plot(q, y_q, 'bo', _q, y__q)
plt.show()
plt.figure()
plt.plot(q, [test_function(x) for x in q], 'r')
plt.show()
```
